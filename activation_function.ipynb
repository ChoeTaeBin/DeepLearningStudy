nn모듈에 activation function이 들어있다.

```python
import torch
from torch import nn
import matplotlib.pylab as plt

x = torch.linspace(-5, 5, 50)

sigmoid = nn.Sigmoid()
tanh = nn.Tanh()
relu = nn.ReLU()

figure = plt.figure(figsize=(12, 4))

ax1 = figure.add_subplot(1, 3, 1)
ax1.plot(x.numpy(), sigmoid(x).detach().numpy())
ax1.set_title('Sigmoid')

ax2 = figure.add_subplot(1, 3, 2)
ax2.plot(x.numpy(), tanh(x).detach().numpy())
ax2.set_title('Tanh')

ax3 = figure.add_subplot(1, 3, 3)
ax3.plot(x.numpy(), relu(x).detach().numpy())
ax3.set_title('ReLU')

plt.show()
```

Softmax

```python
x = torch.tensor([[1, 2, 3],
                  [3, 2, 1.]])
```

```python
m = nn.Softmax(dim = 1) #1차원(열)을 순회하며 softmax 값 계산 할것, 나머지 고정하고 1차원을 순회하며 더하면 1이 됨
print(m(x))
```

```python
m = nn.Softmax(dim = 0) #0차원(행)을 순회하며 softmax 값 계산 할것
print(m(x))
```
